```python
def crete_data_dic(data_filename):
        # Inputs: Name of the data 
    # Output: returns data dic with key as headers and rows for those headers as values
    
    ### BEGIN SOLUTION
    #reading csv file and adding lines to a list
    header = ''
    data = []
    with open(data_filename, 'r') as file:
        for lines in file:
            if not lines.strip():
                continue
            if not header:
                header=lines.split(',')
                continue
            data.append(lines.split(","))
    
    ## creating a dic for headers and data
    data_dic = {}
    for count1, i in enumerate(header):
        inli = []
        for count2,j in enumerate(data):
            inli.append(j[count1].strip())
        data_dic[i.strip()] = inli
    
    return data_dic
    ### END SOLUTION
```

```python
data_filename = "Patient_Characteristics_Survey__PCS__2022.csv"
normalized_database_filename = 'normalized.db'
```

```python
data_dic = crete_data_dic(data_filename)
```

```python
def save_data_as_json(data, filename):
    """
    Write a function whose inputs are a Python dictionary and filename. 
    The function will saves the dictionary as a json file using the filename given. 
    Use the json library. 
    Use these options to correctly format your JSON -- 
    sort_keys=True, indent=4, separators=(',', ': '), ensure_ascii=False. 
    Use this function to save your parsed data as a json file.
    """
    import json
    # BEGIN SOLUTION
    with open(filename, "w") as outfile:
        json.dump(data, outfile,sort_keys=True, indent=4, separators=(',', ': '), ensure_ascii=False)
    # END SOLUTION
```
